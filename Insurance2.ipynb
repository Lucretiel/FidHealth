# Nathan West's Fidessa Insurance Simulator

Hello! This app lets you simulate your costs under the various health insurance plans offered by Fidessa. Let's get started!

First, we load the insurance simulator. Do not edit the following line.

```python
from insurance2 import plan_sim, copay, coinsure, covered, not_covered, generate_services, Service, run_simulation
from pprint import pprint
```

Next, we must fill in the details of the plan. These numbers have been entered from the 2015 Fidessa plans for a single (non-family) employee, and shouldn't require altering. Feel free to edit them as necessary to be in line with your own situation.

```python
POS = plan_sim(
    premium=55,
    employer_contribution=0,  # Yearly contribution to an HSA, for example

    in_network_service_details={
        'deductible': 0,
        'out_of_pocket_max': 1500,
        'PCP': copay(15),
        'specialist': copay(30),
        'preventative': covered(),
        'inpatient': copay(100),
        'outpatient': covered(),
        'urgent': copay(30),
        'ER': copay(75),
        'brand': copay(30),
        'generic': copay(15),
        'non-formulary': copay(50),
    },

    out_of_network_service_details={
        'deductible': 1000,
        'out_of_pocket_max': 1500,
        'PCP': coinsure(20),
        'specialist': coinsure(20),
        'preventative': coinsure(20),
        'inpatient': coinsure(20),
        'outpatient': coinsure(20),
        'urgent': coinsure(20),
        'ER': copay(75),
    })

HDHP = plan_sim(
    premium=15,
    employer_contribution=625,
    
    in_network_service_details={
        'deductible': 1500,
        'out_of_pocket_max': 2500,
        'PCP': covered(),
        'specialist': covered(),
        'preventative': (covered(), True),  # The "True" here indicates that this service ignores the deductible
        'inpatient': covered(),
        'outpatient': covered(),
        'urgent': covered(),
        'ER': covered(),
        'brand': copay(25),
        'generic': copay(10),
        'non-formulary': copay(50),
    },

    out_of_network_service_details={
        'deductible': 2500,
        'out_of_pocket_max': 5000,
        'PCP': coinsure(20),
        'specialist': coinsure(20),
        'preventative': coinsure(20),
        'inpatient': coinsure(20),
        'outpatient': coinsure(20),
        'urgent': coinsure(20),
        'ER': covered(),
    })
```

Finally, enter the monthly and yearly services you expect to receive. Each service has a type (specialist, for example), a price (this is the FULL price of the service, before insurance kicks in), and the word True or False, indicating whether or not the service is recieved in-network. For a service that is received multiple times per year, enter it once for each time it will occur under yearly services.

NOTE: For simplicity, the yearly services are all simulated to occur in January. While this is obviously not realistic, all the resultant math is indentical. Keep in mind, though, that this assumes that the full amount of your employer HSA contribution is available year-round. This means you may have to, in the real word, float some of the upfront costs, and reimburse yourself out
of the HSA later.

```python
run_simulation(POS, HDHP, {
    "Monthly prescription": generate_services(
        monthly_services=[
            Service('generic', 100)]),

    "Low cost": generate_services(
        Service('preventative', 75, True),
        Service('specialist', 75, True),
        Service('specialist', 75, True),
        Service('generic', 25, True),
        Service('generic', 25, True),
        Service('brand', 225, True)),
    
    "Medium cost": generate_services(
        Service('ER', 550, True),
        Service('specialist', 75, True),
        Service('specialist', 75, True),
        Service('specialist', 75, True),
        Service('specialist', 75, True),
        Service('generic', 25, True),
        Service('generic', 25, True),
        Service('generic', 25, True),
        Service('generic', 25, True),
        Service('urgent', 800, True)),

    "High cost": generate_services(
        Service('ER', 550, True),
        Service('ER', 550, True),
        Service('ER', 550, True),
        Service('specialist', 380, True),
        Service('specialist', 380, True),
        Service('specialist', 380, True),
        Service('specialist', 380, True),
        Service('specialist', 380, True),
        Service('specialist', 380, True),
        Service('specialist', 380, True),
        Service('specialist', 380, True),
        Service('specialist', 380, True),
        Service('specialist', 380, True),
        Service('non-formulary', 220, True),
        Service('non-formulary', 220, True),
        Service('non-formulary', 220, True),
        Service('inpatient', 8000, True)),

    "Liana": generate_services(
        monthly_services=[
            Service('generic', 200)],
        yearly_services=[
            Service('specialist', 200),
            Service('specialist', 200),
            Service('specialist', 200),
            Service('specialist', 200)])
})
```
